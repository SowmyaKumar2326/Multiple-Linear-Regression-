```python
q1 = toyota['Price'].quantile(0.25)
print(f"The Q1 value is {q1}")
q3 = toyota['Price'].quantile(0.75)
print(f"The Q3 value is {q3}")
IQR = q3 - q1
print(f"The IQR value is {IQR}")
upper = q3 + (1.5*IQR)
print(f"The Upper threshold value is {upper}")
lower = q1 - (1.5*IQR)
print(f"The Lower threshold value is {lower}")
```

```python
outliers = toyota[(toyota['Price']<lower) | (toyota['Price']>upper)]
print(f'The number of outliers in the data is {len(outliers)} out of {len(toyota)}')
percent = (len(outliers)/len(toyota))*100
print(f'The percentage of Price outliers is {round(percent,4)}')
```

# Model-1

```python
y=toyota[['Price']]
print(y.head(3))
```

```python
X=toyota[['Age_08_04','KM','HP','Weight']]
print(X.head(3))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=450)
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
print(model)
```

```python
train_model=model.fit(X_train,y_train)
print(train_model)
```

```python
pred=train_model.predict(X_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```

```python
#Since we have r2_score as 0.83 this is a very good model.
#The accuracy of the model is 83.17%.
#The price of the vehicle is linearly dependent on the Price,Age_08_04,KM,HP and Weight factors.
```

```python
intercept = train_model.intercept_
slope = train_model.coef_
print('Intercept:',intercept)
print("Slope:",slope)
```
